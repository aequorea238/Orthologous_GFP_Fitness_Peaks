```python
def get_index_of_dispersion(mu, sigma):
    variance = sigma*sigma
    dispersion_index = variance/mu
    return dispersion_index
```

```python
def get_replicates_info_weighted(data, indexer_column, brightness_column, cellcount_column):
    data = data.copy()
    data[indexer_column] = data[indexer_column].replace(np.nan, 'wt')
    
    # get number of replicates for each genotype (different barcodes or different machines)
    data['i exist'] = True
    data = data.reset_index().set_index(indexer_column)
    data['n_replicates'] = data.groupby(data.index)['i exist'].sum()[data.index]
    del data['i exist']
    data = data.reset_index()
    print('Number of replicates has been determined.')
    
    # get mean brightness of each genotype, weighted by the cell counts of the different replicates
    x = data.groupby(data[indexer_column]).apply(lambda x: np.average(x[brightness_column], weights=x[cellcount_column]*100))
    data['replicates_mean_brightness'] = data[indexer_column].map(x)
    print('Brightness means for each genotype have been determined.')

    # get stdev of each genotype, weighted by the cell counts of the different replicates
    y = data.groupby(data[indexer_column]).apply(lambda x: DescrStatsW(x[brightness_column], weights=x[cellcount_column]*100).std)
    data['replicates_stdev_weighted'] = data[indexer_column].map(y)
    l = list(data['replicates_stdev_weighted'])
    data['replicates_stdev_weighted'] = [x if type(x)==np.float64 else 0 for x in l]
    print('St. dev. for each genotype have been determined.')
    
    # get total cell count of each genotype (pooled replicates)
    z = data.groupby(data[indexer_column]).apply(lambda x: sum(x[cellcount_column]))
    data['total_cell_count'] = data[indexer_column].map(z)
    print('Total cell counts have been determined.')
    
    return data
```

```python
def import_data(gene):
    A = pd.read_csv(os.path.join(data_folder, 'unfiltered_barcodes_to_brightness',
                gene+'__unfiltered_barcodes_to_brightness_machine_A.csv'), index_col=0)
    B = pd.read_csv(os.path.join(data_folder, 'unfiltered_barcodes_to_brightness',
                gene+'__unfiltered_barcodes_to_brightness_machine_B.csv'), index_col=0)
    return pd.concat([A,B])
```

### Discard low-quality or unrealiable genotypes (low cell count, abnormal distribution, lack of replicates)

```python
def plot_misgated(data, n_replicates, stdev_replicates, cellcount, brightness_column, brightness_threshold,
                 stdev_column): #, mean_gate_borders):
    df = data.reset_index().copy()
    nt_wt_mask = df['aa_genotype_pseudo']=='wt'
    n_replicates_mask = df['n_replicates'] >= n_replicates
    replicates_stdev_mask = df[stdev_column] <= stdev_replicates
    total_cell_count_mask = df['total_cell_count'] >= cellcount
    column_brightness = brightness_column
    column_count = 'total_cell_count'

    # for negative controls, i.e. genotypes expected not to be bright,
    # chromophore mutantions are more reliable than stop codons (which can be ignored in E coli)
    # or frameshifts (which might occur towards the end of the protein, etc.)
    frameshift_mask = df['nt_genotype'].str.contains('\.')
    nonsense_mask = (df['aa_genotype_pseudo'].str.contains('\*', na=False)) & (~df['aa_genotype_pseudo'].str.contains('\*238',na=True))
    killing_mutations_mask = (df['aa_genotype_pseudo'].str.contains('|'.join(chromomuts['pseudopos']), na=False)) & (~frameshift_mask) & (~nonsense_mask)

    # check how many genotypes are left after filtering under the set conditions
    masks = (replicates_stdev_mask) & (n_replicates_mask) & (total_cell_count_mask)
    laa = len(set(df[masks]['aa_genotype_pseudo']))
    lnt = len(set(df[masks]['nt_genotype']))
    print('There are %s nucleotide genotypes, and %s amino acid genotypes left after these filtering conditions \n' % (lnt, laa))
        
    # plot fitness of wild-types (positive controls) and chromophore mutants (negative controls)
    plt.figure(figsize=[20,3], dpi=200)
    plt.yscale('log')
    plt.ylim(1,10**6)
    plt.xlabel('Fluorescence')
    plt.axvline(x=brightness_threshold, color='k', linestyle='--', alpha=0.5)

    test = df[(nt_wt_mask) & masks]
    
    plt.scatter(test[column_brightness], test[column_count], color='limegreen', alpha=0.1, label='Wildtypes')
    print('Fraction of WT in the dark gates: ', len(test[test[column_brightness]<brightness_threshold])/len(test))
    print('(%s out of %s WT protein genotypes) \n' % (len(test[test[column_brightness]<brightness_threshold]), len(test)))

    test = df[(killing_mutations_mask) &masks]
    plt.scatter(test[column_brightness], test[column_count], color='crimson', alpha=0.1, label='Chromomuts')
    print('Fraction of chromomuts in the bright gates: ', len(test[test[column_brightness]>brightness_threshold])/len(test))
    print('(%s out of %s chromomut genotypes) \n' % (len(test[test[column_brightness]>brightness_threshold]), len(test)))
    
    plt.legend()
```
